# Questão 1 (vale 6 pontos)
Considerando as funções listadas a seguir faça: **(a)** o gráfico das três funções; **(b)** o passo a passo algébrico, em latex, para encontrar o ângulo de inclinação da reta tangente no ponto $x=5$ para cada das funções:

1. $
f(x) = (x^2 - 4)(x^3 - 8x-1)
$
2. $
f(x) = \sin(x^3)\cos(x)
$
3.$
f(x) = x^2e^{-2x}\ln(x)
$

Se você usar o sympy, terá que fazer com que ele imprima todos os passos necessários para resolver cada uma das operações, regras, ou tabelas que aplicar para resolver o problema.

Em todas as questões você pode usar apenas os métodos, regras, técnicas e algoritmos que utilizamos na nossa disciplina, em sala, ou no material de apoio.


```python
import numpy as np
import matplotlib.pyplot as plt
import sympy as sp
from IPython.display import display, Math

def f(x):
    return (x**2 - 4) * (x**3 - 8*x - 1)

def f_derivative(x):
    return 5*x**4 - 36*x**2 - 2*x + 32

#Valores para o eixo x
x = np.linspace(-5, 5, 1000)

# Valores para o eixo y
y = f(x)

# Gráfico da função
plt.plot(x, y, label='f(x)')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Gráfico da função f(x)')
plt.legend()
plt.grid(True)

# Ajustar a escala do eixo y
plt.ylim([-200, 200])

plt.show()

# Cálculo do ângulo de inclinação da reta tangente em x=5
x_point = 5
slope = f_derivative(x_point)
theta = np.arctan(slope)
theta_deg = np.degrees(theta)


theta = sp.atan(slope)
display(Math(r"f'(x) = (x^3 - 8x - 1) + (x^2 - 4)(3x^2 - 8)"))
display(Math(r"f'(5) = 5 \cdot 5^4 - 36 \cdot 5^2 - 2 \cdot 5 + 32"))


display(Math(r"\theta = \arctan\left(\frac{5 \cdot 5^4 - 36 \cdot 5^2 - 2 \cdot 5 + 32}{1}\right)"))
print(f'O ângulo de inclinação da reta tangente em x=5 é: {theta_deg} graus')

```

```python
import numpy as np
import matplotlib.pyplot as plt
import sympy as sp
from IPython.display import display, Math

def f(x):
    return np.sin(x**3) * np.cos(x)

def f_derivative(x):
    return (3 * x**2) * np.cos(x**3) * np.cos(x) - np.sin(x**3) * np.sin(x)

# Valores para o eixo x
x = np.linspace(-5, 5, 1000)

# Valores para o eixo y
y = f(x)

# Gráfico da função
plt.plot(x, y, label='f(x)')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Gráfico da função f(x)')
plt.legend()
plt.grid(True)

plt.show()

# Cálculo do ângulo de inclinação da reta tangente em x=5
x_point = 5
slope = f_derivative(x_point)
theta = np.arctan(slope)
theta_deg = np.degrees(theta)

display(Math(r"f'(x) = (3x^2)\cos(x^3)\cos(x) - \sin(x^3)\sin(x)"))

display(Math(r"f\left(x\right)=\int \sin \left(x^3\right)\cos \left(x\right)dx"))
print(f'O ângulo de inclinação da reta tangente em x=5 é: {theta_deg} graus')
```

```python
import numpy as np
import matplotlib.pyplot as plt
import sympy as sp
from IPython.display import display, Math

def f(x):
    return x**2 * sp.exp(-2*x) * sp.ln(x)

def f_derivative(x):
    return 2*x*sp.exp(-2*x) * sp.ln(x) + x**2 * (-2) * sp.exp(-2*x) * (1/x) + x**2 * sp.exp(-2*x) * (1/x)

# Valores para o eixo x
x = np.linspace(0.01, 100, 1000)

# Valores para o eixo y
y = [f(val).evalf() for val in x]

# Gráfico da função
plt.plot(x, y, label='f(x)')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Gráfico da função f(x)')
plt.legend()
plt.grid(True)

plt.show()

# Cálculo do ângulo de inclinação da reta tangente em x=5
x_point = 5
slope = f_derivative(x_point).evalf()
slope_float = float(slope)
theta = np.arctan(slope_float)
theta_deg = np.degrees(theta)

display(Math(r"\frac{d}{dx}(x^2 e^{-2x} \ln(x))"))
display(Math(r"\frac{d}{dx}(x^2 e^{-2x} \ln(x)) = 2xe^{-2x}\ln(x) + x^2(-2)e^{-2x}\left(\frac{1}{x}\right) + x^2 e^{-2x}\left(\frac{1}{x}\right)"))
print(f'O ângulo de inclinação da reta tangente em x=5 é: {theta_deg} graus')
```

# Questão 2 (vale 2 pontos)
Uma empresa espacial está tentando descobrir o ponto mínimo da órbita de um satélite cuja função orbital é dada por: 
$$f(x)=7\sin(x)-2\cos(x)$$

Encontre o primeiro ponto de mínimo no intervalo $-10<=x<=10$.

Como sempre, você pode usar qualquer ferramenta disponível na internet para lhe auxiliar a resolver este problema. Lembre-se, contudo, que o sympy é uma boa alternativa. 

Em todas as questões você pode usar apenas os métodos, regras, técnicas e algoritmos que utilizamos na nossa disciplina, em sala, ou no material de apoio. Lembrem-se estamos limitados as bibliotecas sympy, numpy e matplotlib.

Para resolver esta questão você precisará de, no minimo, 4 casas decimais de precisão depois da vírgula. 

```python
import sympy as sp
import numpy as np
import matplotlib.pyplot as plt

def f(x):
    return 7 * sp.sin(x) - 2 * sp.cos(x)

# Definir a variável simbólica x
x = sp.Symbol('x')

# Calcular a derivada da função
f_prime = sp.diff(f(x), x)

# Encontrar os pontos críticos (onde a derivada é igual a zero ou indefinida)
critical_points = sp.solve(sp.Eq(f_prime, 0), x)

# Filtrar os pontos críticos que estão dentro do intervalo [-10, 10]
pt = [p for p in critical_points if -10 <= p <= 10]

# Calcular os valores da função nos pontos críticos
values = [f(p) for p in pt]

# Encontrar o ponto de mínimo
min_index = np.argmin(values)
x_min = pt[min_index]
y_min = values[min_index]

print(f'O ponto de mínimo ocorre em x = {x_min:.4f} com f(x) = {y_min:.4f}')
```

# Questão 3 (vale 2 pontos)
Considerando a função $f(x,y) = \sin(xy) + \cos(x+2y)$, no intervalo $0\leq x\leq 5 \space\space e \space\space 0\leq y\leq 5$ trace, usando o plotly o gráfico desta função e marque os pontos de máximo e mínimo neste intevalo.

Lembre-se o sympy é sempre uma opção e você pode usar qualquer técnica, algoritmo, regra, ou tabela, que tenha sido usada na matéria. E que precisa de apenas uma casa decimal de precisão. Uma dica importante para a solução deste problema é a função **np.unravel_index**


```python
import numpy as np
import matplotlib.pyplot as plt
import sympy as sp

#funçao (x,y)
def f(x, y):
    return np.sin(x * y) + np.cos(x + 2 * y)


x = np.linspace(0, 5, 100)
y = np.linspace(0, 5, 100)
X, Y = np.meshgrid(x, y)
Z = f(X, Y)

max_idx = np.unravel_index(np.argmax(Z), Z.shape)
min_idx = np.unravel_index(np.argmin(Z), Z.shape)

#plotagem
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z, cmap='viridis')

#pontos minimos e maximos
ax.scatter(x[max_idx[1]], y[max_idx[0]], Z[max_idx], color='red', label='Ponto de máximo')
ax.scatter(x[min_idx[1]], y[min_idx[0]], Z[min_idx], color='blue', label='Ponto de mínimo')

ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('f(x, y)')
ax.set_title('Gráfico da função f(x, y)')
ax.legend()

#exibiçao do grafico
plt.show()
```
